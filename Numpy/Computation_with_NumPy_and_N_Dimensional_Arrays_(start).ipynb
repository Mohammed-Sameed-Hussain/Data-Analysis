### New Way (The default_rng you read about):

Concept: Imagine buying a specific "Random Number Machine" (the Generator) just for your specific experiment.

- `rng = np.random.default_rng()`: This line buys the machine.

- `rng.random((3, 3, 3))`: This line asks that specific machine to give you numbers.

Why use the confusing New Way? Control. If you are simulating a particle path and want exactly the same "random" path every time you run the code (so you can debug it), you give the machine a "seed" (an ID number):

```python
rng = np.random.default_rng()
matrix_3d = rng.random((3, 3, 3))
print(matrix_3d)
```

```python
rng = np.random.default_rng(seed=42) 
# Now 'rng' will produce the exact same sequence of numbers forever.
rng.random((3,3,3))
```

# Broadcasting and Scalars


Now, oftentimes you'll want to do some sort of operation between an array and a single number. In mathematics, this single number is often called a scalar. For example, you might want to multiply every value in your NumPy array by 2.

In order to achieve this result, NumPy will make the shape of the smaller array - our scalar - compatible with the larger array. This is what the documentation refers to when it mentions the term "broadcasting".

The same rules about 'expanding' the smaller ndarray hold true for 2 or more dimensions. We can see this with a 2-Dimensional Array:

```python
array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])
print(array_2d +10)
print(array_2d*5)
```
